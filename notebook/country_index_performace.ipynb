```python
import pandas as pd
import numpy as np
import sys 
import os
```

```python
DATA_FOLDER = '../data/raw'
```

## Indices

```python
peace_index = pd.read_excel(
    io=os.path.join(DATA_FOLDER, 'GPI-2023-overall-scores-and-domains-2008-2023.xlsx'), 
    skiprows=3, 
    sheet_name='Overall Scores',
    usecols='A:R'
    )

terrorism_index = pd.read_excel(
    io=os.path.join(DATA_FOLDER, 'GTI-2023-Overall-scores-2011-2022.xlsx'),
    skiprows=6, 
    sheet_name='Overall Scores',
    usecols='A:X'
)

ecology_index = pd.read_excel(
    io=os.path.join(DATA_FOLDER, 'ETR_Public_Release_Data_2023.xlsx'),
    skiprows=4, 
    sheet_name='Overall Scores',
)

```

## Combined 

```python
outlook = peace_index.merge(terrorism_index, left_on = ['Country','iso3c'], right_on = ['Country','Iso3c'], how = 'outer', suffixes= ("_peace","_terrorism"))
final = outlook.merge(ecology_index, on = 'Country', how = 'outer', suffixes=("","_ecology"))
```
